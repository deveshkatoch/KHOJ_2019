```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0, 6, 0, 20])
plt.grid()
plt.show()
```

```python
x_axis_array=[]
y_axis_array=[]

for x in range(-50,50,1):    
    y=x**2+2*x+1
    x_axis_array.append(x)
    y_axis_array.append(y)
plt.plot(x_axis_array,y_axis_array)
plt.show()
```

```python
import math
%matplotlib notebook
x_axis_array=[]
y_axis_array=[]

for x in range(-50,50,1):    
    y=math.sin(x)
    x_axis_array.append(x)
    y_axis_array.append(y)
plt.plot(x_axis_array,y_axis_array)
plt.show()
```
